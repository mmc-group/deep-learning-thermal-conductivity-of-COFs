```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
import numpy as np
from matplotlib import cm
from scipy.integrate import simpson
from matplotlib.colors import PowerNorm
from matplotlib.colors import LinearSegmentedColormap

cof_names = ['linker93_C_linker30_C_hcb_relaxed','linker93_C_linker25_C_hcb_relaxed','linker91_C_linker76_C_car_relaxed', \
             'linker91_C_linker62_C_car_relaxed','linker92_C_linker35_C_tts-a_relaxed','linker92_C_linker18_C_tts-a_relaxed',\
             'linker93_C_linker19_C_hca_relaxed','linker93_C_linker20_C_hca_relaxed']

def corrfunc(x, y, **kwargs):
    # corr = np.corrcoef(x, y)[0, 1]  # Pearson's correlation
    corr = r2_score(x, y)   
    ax = plt.gca()
    ax.annotate(r"$R^2$" + f" = {corr:.2f}", 
                xy=(0.15, 0.9), xycoords='axes fraction',
                ha='center', va='center', fontsize=10)
```

**Comparison of thermal conductivity of COFs by MD simulations without electrostatic interactions vs. with
electrostatic interactions**

```python
dataset=pd.read_csv('data/TC-data.csv')
charge_data = pd.read_csv('data/charge/charge-data.csv')

compare_charge = pd.merge(charge_data, dataset, how='left', left_on='cof', right_on='COF ID')
corrfunc(compare_charge['k_avg electrostatic'], compare_charge['k_avg [W/(mK)]'])
plt.scatter(compare_charge['k_avg electrostatic'], compare_charge['k_avg [W/(mK)]'],s=5,c='red')
plt.plot([0,4],[0,4],c='grey',linestyle='--')
plt.xlim([0,2.2])
plt.ylim([0,2.2])
plt.xlabel(r'$\mathrm{\kappa \ [W\ m^{-1}\ K^{-1}]}$'+' with electrostatic interactions')
plt.ylabel(r'$\mathrm{\kappa \ [W\ m^{-1}\ K^{-1}]}$'+' without electrostatic interactions')
plt.gca().set_aspect(1)
```

**VDOS plots**

```python
# Normalize the VDOS spectra
def normalize_vdos(vdos_data):
    frequencies, vdos = vdos_data['# Frequency (THz)'], vdos_data[' VDOS (THz^-1)']
    vdos_normalized = vdos #/ simpson(vdos, frequencies)
    vdos_positive = np.sqrt(vdos_normalized ** 2)
    return vdos_positive, frequencies

def get_overlap(vdos_list):
    vdos_positive = []
    auc_list = []
    for vdos in vdos_list:
        vdos_i, frequencies = normalize_vdos(vdos)
        vdos_positive.append(vdos_i)
    
    f_combined_min = np.minimum.reduce(vdos_positive)
    f_combined_max = np.maximum.reduce(vdos_positive)
    auc_min = simpson(f_combined_min, frequencies)
    auc_max = simpson(f_combined_max, frequencies)

    return auc_min/auc_max
    # return np.mean(auc_list)

props = dict(boxstyle='square', facecolor='lightgray', alpha=0.8)
plt.rcParams.update({'font.size': 12})
```

**pSED plots**

```python
def truncate_colormap(cmap, min_val=0.0, max_val=1.0, n=100):
    new_cmap = LinearSegmentedColormap.from_list(
        'trunc({n},{a:.2f},{b:.2f})'.format(n=cmap.name, a=min_val, b=max_val),
        cmap(np.linspace(min_val, max_val, n)))
    return new_cmap

hot_trunc = truncate_colormap(plt.cm.hot, 0.0, 1.)

def plot(file, vmin=None, vmax=None, title=None):
    sed = np.loadtxt(f'{file}/cof.pSED')[:2000, :]
    sed = np.log(sed)
    # vmin = np.percentile(sed, 1) 
    # vmax = np.percentile(sed, 99)
    print(np.percentile(sed, 0.1), np.percentile(sed, 99.9))

    fig, ax = plt.subplots()
    fig.set_size_inches(3, 4.5, forward=True)
    fig.tight_layout(pad=5)

    # im = ax.imshow(sed, cmap='afmhot', interpolation='hamming', aspect='auto', origin='lower')#,
    #         #   vmin=vmin, vmax=vmax)

    im = ax.imshow(
    sed,
    cmap=hot_trunc,
    interpolation='hamming',
    aspect='auto',
    origin='lower',
    vmin=vmin,
    vmax=vmax)
    # norm=PowerNorm(gamma=0.8, vmin=vmin+1, vmax=vmax-1)  # Adjust gamma < 1 to emphasize low intensities
        # )

    ax.set_yticks([0, sed.shape[0]])
    ax.set_yticklabels(['0', '1'])
    ax.set_xticks([0, sed.shape[1]])
    ax.set_xticklabels([r'$\Gamma$', 'X'])
    # ax.set_xlabel(r'$\bf{q}$')
    ax.set_ylabel('Frequency (THz)')

    fig.subplots_adjust(right=0.85)
    cbar_ax = fig.add_axes([0.88, 0.208, 0.04, 0.625])
    fig.colorbar(im, cax=cbar_ax,shrink=0.5)
    cbar_ax.set_ylabel(r'log($\phi$($\bf{q}$, $\omega$))')
    plt.title(title)

    plt.show()


plot('data/pSED/1d' ,-10.177924398237888, -8.351678752938664, title=cof_names[0])
plot('data/pSED/1nd', -10.177924398237888 ,-8.351678752938664, title=cof_names[1])
plot('data/pSED/2d', -9.220390707829685, -7.784825297703011, title=cof_names[2])
plot('data/pSED/2nd', -9.220390707829685, -7.784825297703011, title=cof_names[3])
plot('data/pSED/3d', -8.149083869851841, -6.939245946073508, title=cof_names[4])
plot('data/pSED/3nd',-8.149083869851841, -6.939245946073508, title=cof_names[5])
plot('data/pSED/4d', -8.294049640102028, -7.223836825955616, title=cof_names[6])
plot('data/pSED/4nd', -8.294049640102028,-7.223836825955616, title=cof_names[7])
```
